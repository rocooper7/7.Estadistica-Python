```python
#Distribuciones Discretas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Distribuci√≥n de Bernulli
from scipy.stats import bernoulli     #Se puede sacar tambien con numpy (np.random.normal)
p = 0.3      #Probabilidad de exito
data = bernoulli.rvs(p, size = 100)
len(data)
```

```python
import numpy as np
np.unique(data,return_counts=True)
```

```python
import pandas as pd
pd.Series(data).value_counts()
```

```python
mean, var, skew, kurt = bernoulli.stats(p, moments = 'mvsk')
mean, var, skew, kurt
```

```python
import seaborn as sns

ax = sns.distplot( data, bins = 30, kde = False, color = 'blue')    #Displot == histogram de seaborn
ax.set( xlabel = 'Distribucion Bernoulli', ylabel = 'Frecuencia' )
ax.axvline(x = mean, linestyle = '--', label = 'Media' )
ax.legend();
```

```python
# Distribuci√≥n Binomial
from scipy.stats import binom
p = 0.3
n = 10
data = binom.rvs(p = p, n= n, size = 100)
len(data)
```

```python
data
```

```python
pd.Series(data).value_counts()
```

```python
mean, var, skew, kurt = binom.stats(p = p, n = n , moments = 'mvsk')
mean, var, skew, kurt
```

```python
ax = sns.distplot( data, bins = 30, kde = False, color = 'blue') 
ax.set( xlabel = 'Distribucion binomial', ylabel = 'Frecuencia' )
ax.axvline(x = mean, linestyle = '--', label = 'Media' )
ax.legend();
```

```python
#Distribuci√≥n Binomial Negativa
from scipy.stats import nbinom
p = 0.3
n = 10
data = nbinom.rvs(p = p, n= n, size = 100)
len(data)
data
```

```python
pd.Series(data).value_counts().head()
```

```python
mean, var, skew, kurt = binom.stats(p = p, n = n , moments = 'mvsk')
mean, var, skew, kurt
```

```python
ax = sns.distplot( data, bins = 30, kde = False, color = 'blue') 
ax.set( xlabel = 'Distribucion binomial negativa', ylabel = 'Frecuencia' )
ax.axvline(x = mean, linestyle = '--', label = 'Media' )
ax.legend();
```

```python
from scipy.stats import poisson
lambda_p = 100   #Unidades de tiempo
data = poisson.rvs(mu = lambda_p, size = 100)
len(data)
data
```

```python
pd.Series(data).value_counts().head()
```

```python
mean, var, skew, kurt = poisson.stats(mu =lambda_p, moments = 'mvsk')
mean, var, skew, kurt
```

```python
ax = sns.distplot( data, bins = 30, kde = False, color = 'blue') 
ax.set( xlabel = 'Distribucion poisson', ylabel = 'Frecuencia' )
ax.axvline(x = mean, linestyle = '--', label = 'Media' )
ax.legend();
```

```python
#Distribuciones Continuas
#Distribuci√≥n Uniforme
from scipy.stats import uniform
data = uniform.rvs(size = 100000000)
len(data)
```

```python
mean, var, skew, kurt = uniform.stats( moments = 'mvsk')
mean, var, skew, kurt
```

```python
ax = sns.distplot( data, bins = 500, kde = False, color = 'blue') 
ax.set( xlabel = 'Distribucion uniform 0 -1 ', ylabel = 'Frecuencia' )
ax.axvline(x = mean, linestyle = '--', label = 'Media' )
ax.legend();
```
