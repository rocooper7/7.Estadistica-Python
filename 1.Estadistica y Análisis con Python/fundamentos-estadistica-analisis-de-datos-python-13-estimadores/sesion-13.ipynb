```python
import sklearn 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
from scipy.stats import norm

x1 = 3   #Muestra

mu1 = 4   #Hipótesis     Muestra poblacional 
mu2 = 7   
sigma = 1   #Desviación Estandar 

p_muestra = norm.pdf(x1, mu2, sigma)     #probabilidad de la muestra
p_muestra
```

```python
from scipy.stats import norm     #Probabilidad Conjunta

x1 = 3
x2 = 10

mu1 = 4
mu2 = 7
sigma = 1

p_muestra = norm.pdf(x1, mu1, sigma)*norm.pdf(x2, mu1, sigma)
p_muestra
```

```python
muestra_10 = norm.rvs( 5  , sigma, size = 10)
muestra_10
```

```python
data1 = norm.rvs( mu1  , sigma, size = 100000)
data2 = norm.rvs( mu2  , sigma, size = 100000)
data1,data2
```

```python
ax = sns.distplot(data1, bins = 50, color = 'blue', kde = False)    #kde se usa para eliminar la probabilidad 
ax.set(xlabel = 'Distribucion normal mu1', ylabel = 'Frecuencia')

ax = sns.distplot(data2, bins = 50, color = 'red',  kde = False)
ax.set(xlabel = 'Distribucion normal mu2', ylabel = 'Frecuencia');
```
