```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

from scipy.stats import expon
from scipy.stats import uniform
```

```python
poblacion = pd.DataFrame()
poblacion['numbers'] = expon.rvs(40, size = 100000 )
poblacion
```

```python
from scipy import stats
y = poblacion.numbers
stats.mode(y)#[0] 
```

```python
poblacion['numbers'].hist( bins = 100);    #Graficando con pandas 
```

```python
poblacion.sample(n=3).mean().values[0]
```

```python
muestra_promedio = []

tamano = 5000

for i in range(0, tamano):
    muestra_promedio.append( poblacion.sample(n=100).mean().values[0])
```

```python
fig, ax = plt.subplots()
ax.hist(muestra_promedio, bins = 50, alpha = 0.5)
ax.set_xlabel('Valor promedio')
ax.set_ylabel('Frecuencia')
ax.grid()
```
